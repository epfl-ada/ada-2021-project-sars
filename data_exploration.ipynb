```python
import pandas as pd
import seaborn as sns
```

```python
# Read the first 10000 quotations and store them in a datafram
with pd.read_json('Quotebank/quotes-2020.json.bz2', lines=True, compression='bz2', chunksize=1000000) as df_reader:
    for chunk in df_reader:
        test_df = chunk
        
        for i in range(len(chunk)):
            a = len(chunk['urls'][i])
            b = chunk['numOccurrences'][i]
            if a != b:
                print(f'data incoherence found at row {i} {a} != {b}')
        break

test_df.head()   
```
